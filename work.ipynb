```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
xNum=pd.DataFrame(df[['Age','SibSp','Fare','Survived']])
xNum.describe()
```

```python
print(xNum.corr())

# Heatmap of the Correlation Matrix
f, axes = plt.subplots(1, 1, figsize=(20, 20))
sb.heatmap(xNum.corr(), vmin = -1, vmax = 1, linewidths = 1,
           annot = True, fmt = ".2f", annot_kws = {"size": 18}, cmap = "RdBu")
```

```python
sb.pairplot(data = xNum)
```

```python
xCat = pd.DataFrame(df[['Pclass', 'Sex', 'SibSp', 'Parch','Embarked','Cabin']])
xCat.head()
```

```python
xCat['Pclass'] = xCat['Pclass'].astype('category')
xCat['Sex'] = xCat['Sex'].astype('category')
xCat['SibSp'] = xCat['SibSp'].astype('category')
xCat['Parch'] = xCat['Parch'].astype('category')
xCat['Embarked'] = xCat['Embarked'].astype('category')
xCat['Cabin'] = xCat['Cabin'].astype('category')
```

```python
xCat.describe()
```

```python

for i in ['Pclass', 'Sex', 'SibSp', 'Parch','Embarked','Cabin']:
    sb.catplot(y = i, data = xCat, kind = "count", height = 8)
```
